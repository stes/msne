# Comparison of different neuron models

Visualization of the models implemented in the ``neurons.py`` module.

```python
import numpy as np
import seaborn as sns
from neurons import hodgkin_huxley, hodgkin_huxley_simple, fitzhugh_nagumo, leaky_integrate
from pylab import *
sns.set_style("white")
sns.set_context("notebook")
%matplotlib inline
```

```python
import scipy.signal

dt = 0.01
nb_timesteps = 10000
nb_neurons = 1

tt = np.linspace(0,dt*nb_timesteps,nb_timesteps)
I = np.zeros((nb_timesteps, nb_neurons))
I[nb_timesteps//10:nb_timesteps//10*9] = 20

models = [hodgkin_huxley, hodgkin_huxley_simple, fitzhugh_nagumo, leaky_integrate]
names  = ["Hodgkin Huxley", "Simplified Hodgkin Huxley Model", "FitzHugh Nagumo", "Leaky Integrate and Fire"]

figure(figsize=(15,10))
for i, model in enumerate(models):
    subplot(2,2,i+1)
    
    u, gating_params = model(tt, I)
    plot(tt, u[:,0])    
    
    xlabel("Time [ms]")
    ylabel("Voltage [mV]")
    title(names[i])
    ylim([-80,50])

show()
```
