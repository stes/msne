## Data Analysis

Below are some plots of data recorded from the actual EMG amplifier.

```python
import pandas as pd
import numpy as np
from pylab import *
%matplotlib inline
```

```python
finger_data = pd.read_csv("data/finger.csv")
tt1 = finger_data["Time (s)"]
uu1 = finger_data["Channel 1 (V)"]

thumb_data = pd.read_csv("data/thumb.csv")
tt2 = thumb_data["Time (s)"]
uu2 = thumb_data["Channel 1 (V)"]


plot(tt1, uu1)
title("Finger Movement")
xlabel("Time [s]")
ylabel("Volate [V]")
show()
plot(tt2, uu2)
title("Thumb Movement")
xlabel("Time [s]")
ylabel("Volate [V]")
show()
```

```python
1 = uu1.mean()
2 = uu2.mean()

sigma = (uu1 - 1)**2
plot(tt1, sigma)
title("Finger Movement")
xlabel("Time [s]")
ylabel("Variance [V]")
show()

sigma = (uu2 - 2)**2
title("Thumb Movement")
xlabel("Time [s]")
ylabel("Variance [V]")
plot(tt2, sigma)
show()
```
